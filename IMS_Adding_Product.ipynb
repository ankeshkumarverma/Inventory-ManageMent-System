```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

# ADDING PRODUCT IN **INVENTORY**

```python
prod_id = str(input("Enter Product Id: "))
#print("This is new Product which is going to be added in Inventory..")
name = str(input("Please Enter Product Name: "))
pr = int(input("Enter Price: "))
qn = int(input("Enter Quantity: "))
brand = str(input("Enter Brand of PProduct: "))
catego = str(input("Enter Cotegory of Product: "))
color = str(input("Enter Lookinf-Color of the Product: "))
rt = int(input("Prevous Rating Of Product: Choose anybetween 1 to 5: "))
typ = str(input("Enter Product Type: "))
size = float(input("Enter Size of the Product in which it belongs : "))
made_in = str(input("Enter Product-Manufactured Country: "))
ideal_for = str(input("Product is ideal for Choose one-> Men (or) Women (or) Child (or) Young: "))
for i in record.keys():
  if ( prod_id == i ):
    qn = qn + record[i]['qn']
record[prod_id] = {'name': name, 'pr': pr, 'qn': qn, 'brand': brand, 'category': catego, 'color': color, 'customer_rating': rt, 'type': typ, 'size': size, 'made_in': made_in, 'ideal_for': ideal_for}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

# **Deleting a Record**

```python
del_id = input("Enter the id of Product which you want to delete: ")
del record[del_id]
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
